```python
with open('./log_gui_restart.log') as f:
    data = f.readlines()
```

```python
data = data[100:]
```

```python
data = map(lambda s: s.strip(), data)
```

```python
data = [_ for _ in data if _ is not '' and '_dcue' not in _]
```

```python
local_time = np.array(data[::2]).astype(float)
```

```python
local_time = local_time[:-1]
local_time
```

```python
plot(local_time[:],'.')
```

```python
np.mean(local_time), std(local_time)
```

```python
hist(local_time, bins=20);
```

```python
jovian_time = []
for _time in data[1::2]:
    jovian_time.append(int(_time[2:-3].split(',')[0]))
```

```python
jovian_time = np.array(jovian_time)
```
